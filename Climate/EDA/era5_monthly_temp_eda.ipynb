## ERA5 Average Monthly Temperature

```python
import pandas as pd
import numpy as np
import xarray as xr 
import proplot as plot 
import matplotlib.pyplot as plt
import cftime

from tqdm import tqdm
```

```python
era5path = "/home/johnyannotty/NOAA_DATA/ERA5/era5_land_avg_mon_tas_1990-2023/"
era5 = xr.open_dataset(era5path+"data.nc", decode_times=True, use_cftime=True)
```

```python
# Plot avg temperatures by month
fig, ax = plt.subplots(2,2,figsize = (15,10)) 
xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(1990, 1, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[0][0], vmin = 200, vmax = 325)

xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(2000, 1, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[0][1], vmin = 200, vmax = 325)

xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(2010, 1, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[1][0], vmin = 200, vmax = 325)

xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(2020, 1, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[1][1], vmin = 200, vmax = 325)

fig.suptitle("Avg January Surface Temperature (K) by Year", size = 20)
fig.tight_layout()
```

```python
# Plot avg temperatures by month
fig, ax = plt.subplots(2,2,figsize = (15,10)) 
xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(1990, 6, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[0][0], vmin = 200, vmax = 325)

xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(2000, 6, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[0][1], vmin = 200, vmax = 325)

xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(2010, 6, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[1][0], vmin = 200, vmax = 325)

xx = np.where(era5["t2m"][:,0,:,:].time.isin(cftime.DatetimeGregorian(2020, 6, 1, 0, 0, 0, 0, 2, 15)))[0][0]
era5["t2m"][xx,0,:,:].plot(cmap = "coolwarm", ax = ax[1][1], vmin = 200, vmax = 325)

fig.suptitle("Avg June Surface Temperature (K) by Year", size = 20)
fig.tight_layout()
```
