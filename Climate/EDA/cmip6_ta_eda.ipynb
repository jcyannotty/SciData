```python
import pandas as pd
import numpy as np
import xarray as xr 
import proplot as plot 
import matplotlib.pyplot as plt
#import cartopy.crs as ccrs
import cftime

from tqdm import tqdm
```

```python
cmip6path = "/home/johnyannotty/NOAA_DATA/CMIP6/"
filename =  "ta_Amon_BCC-CSM2-MR_hist-nat_r1i1p1f1_gn_197001-200912.nc"
dset = xr.open_dataset(cmip6path+filename, decode_times=True, use_cftime=True)
```

```python
# Plot avg temperatures by month
fig, ax = plt.subplots(2,2,figsize = (15,10)) 
dset["ta"][0][0].plot(cmap = 'coolwarm', ax = ax[0][0], vmin = 230, vmax = 325)
dset["ta"][240][0].plot(cmap = 'coolwarm', ax = ax[0][1], vmin = 230, vmax = 325)
dset["ta"][360][0].plot(cmap = 'coolwarm', ax = ax[1][0], vmin = 230, vmax = 325)
dset["ta"][468][0].plot(cmap = 'coolwarm', ax = ax[1][1], vmin = 230, vmax = 325)
fig.suptitle("Avg January Temperature (K) by Year at Fixed Pressure", size = 20)
fig.tight_layout()
#dset["ta"][0][0].values
```
